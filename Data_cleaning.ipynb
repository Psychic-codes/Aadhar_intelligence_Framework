```python
import re
import unicodedata

def normalize_district(name):
    if pd.isna(name):
        return name

    name = str(name)

    # Unicode normalization
    name = unicodedata.normalize("NFKD", name)

    # Lowercase
    name = name.lower()

    # Replace common separators
    name = name.replace('&', 'and')
    name = re.sub(r'[−–—]', '-', name)

    # Remove text inside brackets
    name = re.sub(r'\(.*?\)', '', name)

    # Remove special symbols, asterisks, question marks
    name = re.sub(r'[*?]', '', name)

    # Fix multiple spaces
    name = re.sub(r'\s+', ' ', name).strip()

    # Standardize common aliases
    replacements = {
        "bara banki": "barabanki",
        "bardhaman": "purba bardhaman",
        "hugli": "hooghly",
        "punch": "poonch",
        "k.v. rangareddy": "ranga reddy",
        "rangareddi": "ranga reddy",
        "y. s. r": "ysr kadapa",
        "cuddapah": "ysr kadapa",
        "anantapuramu": "anantapur",
        "north cachar hills": "dima hasao",
        "the nilgiris": "nilgiris",
        "south salmara mankachar": "south salmara-mankachar",
        "kamrup metro": "kamrup metropolitan",
        "sas nagar": "sahibzada ajit singh nagar",
        "mohali": "sahibzada ajit singh nagar",
        "bijapur(kar)": "vijayapura",
        "osmanabad": "dharashiv",
        "aurangabad": "chhatrapati sambhajinagar",
        "leh (ladakh)": "leh",
        "bandipore": "bandipora",
        "khordha": "khordha",
        "jajapur": "jajpur",
        "nabarangapur": "nabarangpur",
        "jagatsinghapur": "jagatsinghpur",
        "arvalli": "aravalli",
        "chhota udepur": "chhota udaipur",
        "surendra nagar": "surendranagar",
        "medchalmalkajgiri": "Medchal-Malkajgiri",
        "ysr kadapa": "Ysr Kadapa",
        "n t r": "Ntr",
        "gaurela pendra marwahi": "Gaurela-Pendra-Marwahi",
        "angul": "Angul",
        "anugul": "Angul",
        "lahul and spiti": "Lahaul And Spiti",
        "manendragarhchirmiribharatpur": "Manendragarh-Chirmiri-Bharatpur",
        "Bangalore Rural": "Bengaluru Rural",
        "Bangalore South": "Bengaluru South",
        "Kancheepuram": "Kanchipuram"
    }

    if name in replacements:
        name = replacements[name]

    # Title case for final output
    return name.title()


# Apply normalization
df['district'] = df['district'].apply(normalize_district)

# Check uniqueness
df['district'].nunique()
```

```python
import re
import unicodedata

def normalize_district(name):
    if pd.isna(name):
        return name

    name = str(name)

    # Unicode normalization
    name = unicodedata.normalize("NFKD", name)

    # Lowercase
    name = name.lower()

    # Replace common separators
    name = name.replace('&', 'and')
    name = re.sub(r'[−–—]', '-', name)

    # Remove text inside brackets
    name = re.sub(r'\(.*?\)', '', name)

    # Remove special symbols, asterisks, question marks
    name = re.sub(r'[*?]', '', name)

    # Fix multiple spaces
    name = re.sub(r'\s+', ' ', name).strip()

    # Standardize common aliases
    replacements = {
        "bara banki": "barabanki",
        "bardhaman": "purba bardhaman",
        "hugli": "hooghly",
        "punch": "poonch",
        "k.v. rangareddy": "ranga reddy",
        "rangareddi": "ranga reddy",
        "y. s. r": "ysr kadapa",
        "cuddapah": "ysr kadapa",
        "anantapuramu": "anantapur",
        "north cachar hills": "dima hasao",
        "the nilgiris": "nilgiris",
        "south salmara mankachar": "south salmara-mankachar",
        "kamrup metro": "kamrup metropolitan",
        "sas nagar": "sahibzada ajit singh nagar",
        "mohali": "sahibzada ajit singh nagar",
        "bijapur(kar)": "vijayapura",
        "osmanabad": "dharashiv",
        "aurangabad": "chhatrapati sambhajinagar",
        "leh (ladakh)": "leh",
        "bandipore": "bandipora",
        "khordha": "khordha",
        "jajapur": "jajpur",
        "nabarangapur": "nabarangpur",
        "jagatsinghapur": "jagatsinghpur",
        "arvalli": "aravalli",
        "chhota udepur": "chhota udaipur",
        "surendra nagar": "surendranagar",
        "medchalmalkajgiri": "Medchal-Malkajgiri",
        "ysr kadapa": "Ysr Kadapa",
        "n t r": "Ntr",
        "gaurela pendra marwahi": "Gaurela-Pendra-Marwahi",
        "angul": "Angul",
        "anugul": "Angul",
        "lahul and spiti": "Lahaul And Spiti",
        "manendragarhchirmiribharatpur": "Manendragarh-Chirmiri-Bharatpur",
        "Bangalore Rural": "Bengaluru Rural",
        "Bangalore South": "Bengaluru South",
        "Kancheepuram": "Kanchipuram"
    }

    if name in replacements:
        name = replacements[name]

    # Title case for final output
    return name.title()


# Apply normalization
df6['district'] = df6['district'].apply(normalize_district)

# Check uniqueness
df6['district'].nunique()
```

```python
import re
import unicodedata

# 1. Known district rename / alias mapping
DISTRICT_MAP = {
    "South Twenty Four Parganas": "South 24 Parganas",
    "South 24 Pargana": "South 24 Parganas",
    "East Midnapore": "Purba Medinipur",
    "West Midnapore": "Paschim Medinipur",
    "Jyotiba Phule Nagar": "Amroha",
    "Dist : Thane": "Thane",
    "Medchalâ\x88\x92Malkajgiri": "Medchal-Malkajgiri",
    "Nellore": "Sri Potti Sriramulu Nellore",
    "Aurangabad" :"Chhatrapati Sambhajinagar",
    "Osmanabad": "Dharashiv",
    "Bangalore Rural": "Bengaluru Rural",
    "Bangalore South": "Bengaluru South",
    "Kancheepuram": "Kanchipuram"
}

# 2. Normalization function
def clean_district(text):
    if text is None:
        return None

    text = str(text).strip()

    # remove numeric / garbage
    if text.isdigit() or len(text) < 3:
        return None

    # unicode normalize
    text = unicodedata.normalize("NFKD", text)

    # remove address-like phrases
    if re.search(r"(near|road|colony|cross|thana|hospital|garden)", text, re.I):
        return None

    # remove special characters
    text = re.sub(r"[^A-Za-z0-9\s\.]", "", text).strip()

    return DISTRICT_MAP.get(text, text)

# 3. Apply cleaning
df7["district"] = df7["district"].apply(clean_district)
```
